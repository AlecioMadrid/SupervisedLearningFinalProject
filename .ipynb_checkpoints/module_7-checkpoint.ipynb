```python
# EDA: The first step in the exploratory data analysis is to reduce the feature space size to prevent overfitting and remove any
# significant colinearities. In order to do this we limit the size of the feature space to maximum 10 features and to maximize the 
# accuracy of our model we choose the 10 most significant features. These were calculated using the correlation matrix of the full
# 36 feature data set and then sorting the 10 with the highest correlation to sale price. The 10 features are ['OverallQual', 'GrLivArea', 
# 'GarageCars', 'GarageArea', 'TotalBsmtSF', 'FirstFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt', 'YearRemodAdd'].
# In the analysis in the notebook you can also see the correlation matrix and associated heat map for the 10 most signficant features. 
# The next step was to fit a model with these 10 features and determine if there are any further features that do not meet the p>0.05 
# threshold of significance. Based on the summary plot for the model farther below, it is possible to see that there are two additional
# features that we can remove due to insignificance as determined by the P test. These two features are GarageArea and TotRmsAbvGrd. This
# leaves us with 8 features that have high correlation with sale price and meet the criteria from a p test.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# Read in data set
full_df = pd.read_csv('./data.csv')
```

```python
# Print data shape
full_df.shape
```

```python
# Print dataframe head
full_df.head()
```

```python
# Get dataframe column info
full_df.info()
```

```python
# Get rid of id and non numerical features
df_minus_id = full_df.drop(['Id'], axis=1)
object_cols = df_minus_id.select_dtypes(include=['object']).columns
df_reduced = df_minus_id.drop(columns=object_cols)
df_reduced.info()
```

```python
# Rename 1stFlrSF and 2ndFlrSF
df_reduced.rename(columns={'1stFlrSF': 'FirstFlrSF', '2ndFlrSF':'SecondFlrSF'}, inplace=True)
df_reduced.info()
```

```python
# Get best guess predictors
top = list(df_reduced.corr()['SalePrice'].iloc[:-1].sort_values()[-10:].index)

top_ten = top[::-1]
df_top_10 = df_reduced[top_ten + ['SalePrice']]

print(top_ten)
df_top_10.info()
```

```python
# Get correlation matrix
df_top_10.corr()
```

```python
sns.heatmap(df_top_10.corr())
```

```python
# Let's create a model using the top 10 predictors
res_top_10 = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + GarageArea + TotalBsmtSF + FirstFlrSF + FullBath + TotRmsAbvGrd + YearBuilt + YearRemodAdd', data=df_top_10).fit()
print(res_top_10.summary())
```

```python
# Looks like we should drop TotRmsAbvGrd and GarageArea
df_best_8 = df_top_10.drop(['TotRmsAbvGrd', 'GarageArea'], axis=1)
```

```python
# Get correlation matrix
df_best_8.corr()
```

```python
sns.heatmap(df_best_8.corr())
```

```python
# Let's see what happens when we train a model with smaller and smaller subsets of features
# We will progressively drop the least best guess predictor for each model
res_8 = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF + FullBath + YearBuilt + YearRemodAdd', data=df_reduced).fit()
print(res_8.summary())

res_7 = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF + FullBath + YearBuilt', data=df_reduced).fit()
print(res_7.summary())

res_6 = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF + FullBath', data=df_reduced).fit()
print(res_6.summary())

res_5 = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF', data=df_reduced).fit()
print(res_5.summary())

res_4 = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF', data=df_reduced).fit()
print(res_4.summary())

res_3 = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars', data=df_reduced).fit()
print(res_3.summary())

res_2 = smf.ols('SalePrice ~ OverallQual + GrLivArea', data=df_reduced).fit()
print(res_2.summary())

res_1 = smf.ols('SalePrice ~ OverallQual', data=df_reduced).fit()
print(res_1.summary())
```

```python
# To prepare to run some tests lets get a train test split
X_train, X_test, y_train, y_test = train_test_split(df_best_8.drop('SalePrice', axis=1), df_best_8['SalePrice'], test_size=.2)
train_data = pd.concat([X_train, y_train], axis=1)
test_data = pd.concat([X_test, y_test], axis=1)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Let's retrain all of the models on the training data
# Let's see what happens when we train a model with smaller and smaller subsets of features
# We will progressively drop the least best guess predictor for each model
res_8_train = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF + FullBath + YearBuilt + YearRemodAdd', data=train_data).fit()
print(res_8_train.summary())

res_7_train = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF + FullBath + YearBuilt', data=train_data).fit()
print(res_7_train.summary())

res_6_train = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF + FullBath', data=train_data).fit()
print(res_6_train.summary())

res_5_train = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF + FirstFlrSF', data=train_data).fit()
print(res_5_train.summary())

res_4_train = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars + TotalBsmtSF', data=train_data).fit()
print(res_4_train.summary())

res_3_train = smf.ols('SalePrice ~ OverallQual + GrLivArea + GarageCars', data=train_data).fit()
print(res_3_train.summary())

res_2_train = smf.ols('SalePrice ~ OverallQual + GrLivArea', data=train_data).fit()
print(res_2_train.summary())

res_1_train = smf.ols('SalePrice ~ OverallQual', data=train_data).fit()
print(res_1_train.summary())
```

```python
# Get average house price for reference
average_house_price = np.mean(y_test)
print('avg house price: ' + str(average_house_price))
```

```python
# Get predictions and accuracy on test data for model 8
test_predictions_model_8 = res_8_train.predict(test_data.drop('SalePrice', axis=1))
print(test_predictions_model_8)

rmse_model_8 = np.sqrt(mean_squared_error(y_test, test_predictions_model_8))
print('\nrmse: ' + str(rmse_model_8))
```
